```python
from konlpy.tag import Kkma

analyzer = Kkma()

a = open('../data/train_data.txt', 'r', encoding='utf-8')

test = a.readlines()
a.close()
b = open('../data/dual_data.txt', 'a', encoding='utf-8')

for lines in test:
    for text in lines.strip().split('\t')[1].split('|'):
        if text == '':
            print('pass')
        else:
            try:
                sent = analyzer.pos(text)
                sentence = []
                for word in sent:
                    sentence.append(word[0])
                morpheme = ' '.join(sentence)
                print(morpheme)
                b.write(morpheme+'\n')
            except:
                print('pass2')
                pass
    b.write('<end_of_story>'+'\n')
```

```python
import json
a = open('../data/dual_data.txt', 'r', encoding='utf-8')
test = a.readlines()
a.close()
data = []

for i, line in enumerate(test):
    data_dict = {}
    try:
        next_line = test[i+1].strip()
        this_line = line.strip()
        if this_line == '<end_of_story>' or next_line == '<end_of_story>':
            pass
        else:
            data_dict['input_text'] = this_line
            data_dict['output_text'] = next_line
            data.append(data_dict)
    except:
        print('end')

a = open('../data/dual_data_dict.json', 'w', encoding='utf-8')
a.write(json.dumps(data, indent=2, ensure_ascii=False))
a.close()
```
