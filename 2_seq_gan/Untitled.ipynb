```python
import re
with open('../data/train_data.txt', encoding = 'utf-8') as f:
    texts = f.readlines()
    print(len(texts))

count = 0
stories = list()
for desc in texts[:1]:
    print(desc)
    sentences = list()
    text = desc.split('\t')[1].replace('\n','').split('|')
    #print(text)
    for sent in text:
        if sent == '':
            continue
        sent = re.sub(r"[^ㄱ-힣a-zA-Z0-9]+", ' ', sent)
        if sent:
            sent = analyzer.pos(sent)
        sentences.append(sent)
        count += 1
    stories.append(sentences)
```
