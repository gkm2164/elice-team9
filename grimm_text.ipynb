```python
with open('./grim_new3.txt', encoding = 'utf-8') as f:
    texts = f.readlines()
    cleaned = [line[3:] for line in texts]
```

```python
cleaned = cleaned[1:]
```

```python
len(cleaned)
```

```python
new_cleaned = []
for clean_line in cleaned:
    remove_status = 'no'
    new_clean_line =''
    
    try:
        remove_pre = clean_line.split('\t')
        remove_pre2 = remove_pre[1].split('동화제목 : ')[1]
        remove_pre = remove_pre2.split('|')[0]+'\t'+remove_pre2[6:]
        clean_line = remove_pre
        #print(clean_line)
    except:
        clean_line = clean_line
    
    try:
        remove_pre = clean_line.split('\t')
        remove_pre2 = remove_pre[1].split('(분문 시작)|')[1].split(remove_pre[0]+'|')
        remove_pre = remove_pre[0]+'\t'+remove_pre2[1]
        clean_line = remove_pre
    except:
        clean_line = clean_line
    

    
    for word in clean_line:
        if word == '(':
            remove_status = 'rm'
        elif word == ')':
            remove_status = 'no'
        
        if remove_status == 'no':
            new_clean_line += word
    new_cleaned.append(new_clean_line.replace(')','').replace(' |','|').replace('||','|').replace('\t|','\t').replace('원문→사랑스러운 백설 공주|',''))

    
```

```python
new_cleaned[52][:300]
```

```python
for clean_text in new_cleaned:
    f = open('grimm-split.txt', 'a', encoding = 'utf-8')
    f.write(clean_text)
    f.close()
```
